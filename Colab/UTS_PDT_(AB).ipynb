```python
import pandas as pd
import numpy as np

# Load Data .csv
url = "https://raw.githubusercontent.com/TheresiaTriOktavia/PDT/main/supermarket_sales.csv"
df1 = pd.read_csv(url)
df1
```

```python
# Melihat informasi data (adakah data yang null dan tipe data pada tiap kolom)
df1.info()
```

```python
# Menghapus kolom non numerik
df_clean1 = df1.drop(["Invoice ID", "Branch", "City", "Customer type", "Gender", "Product line", "Date", "Time", "Payment"], axis=1)
df_clean1
```

```python
# Menghitung jumlah (sum) pada tiap kolom
# Mengatur sehingga hanya ada 2 angka di belakang koma
DataSum1 = df_clean1.sum().apply(lambda x: "{:.2f}".format(x))
DataSum1
```

```python
# Menghitung rata-rata (mean) pada tiap kolom
# Mengatur sehingga hanya ada 2 angka di belakang koma
DataMean1 = df_clean1.mean().apply(lambda x: "{:.2f}".format(x))
DataMean1
```

```python
# Menghitung nilai tengah (median) pada tiap kolom
# Mengatur sehingga hanya ada 2 angka di belakang koma
DataMedian1 = df_clean1.median().apply(lambda x: "{:.2f}".format(x))
DataMedian1
```

```python
# Load Data
url = "https://raw.githubusercontent.com/TheresiaTriOktavia/PDT/main/Data_Kursi_Bioskop.csv"
df3 = pd.read_csv(url)
df3.set_index('No Kursi', inplace=True)
df3
```

```python
import pandas as pd

def isi_kursi(df3, nomor_kursi):
    if df3.at[nomor_kursi, 'Keterangan'] == 'Kosong':
        df3.at[nomor_kursi, 'Keterangan'] = 'Terisi'
        print(f"Kursi nomor {nomor_kursi} telah berhasil diisi.\n")
    else:
        print(f"Maaf, kursi nomor {nomor_kursi} sudah terisi.\n")

def kosongkan_kursi(df3, nomor_kursi):
    if df3.at[nomor_kursi, 'Keterangan'] == 'Terisi':
        df3.at[nomor_kursi, 'Keterangan'] = 'Kosong'
        print(f"Kursi nomor {nomor_kursi} telah berhasil dikosongkan.\n")
    else:
        print(f"Maaf, kursi nomor {nomor_kursi} sudah kosong.\n")

def display_data(df3):
    print("Data Kursi Bioskop:\n")
    print(df3)


# Loop untuk mengisi atau mengosongkan kursi
while True:
    print("\nSilakan pilih menu:")
    print("1. Isi kursi")
    print("2. Kosongkan kursi")
    print("3. Tampilkan data kursi")
    print("4. Keluar")
    pilihan = input("Masukkan pilihan menu: ")

    if pilihan == '1':
        print("Silakan masukkan nomor kursi yang ingin Anda isi:")
        nomor_kursi = int(input())
        if nomor_kursi in df3.index:
            isi_kursi(df3, nomor_kursi)
        else:
            print("Nomor kursi tidak valid. Silakan coba lagi.")

    elif pilihan == '2':
        print("Silakan masukkan nomor kursi yang ingin Anda kosongkan:")
        nomor_kursi = int(input())
        if nomor_kursi in df3.index:
            kosongkan_kursi(df3, nomor_kursi)
        else:
            print("Nomor kursi tidak valid. Silakan coba lagi.")

    elif pilihan == '3':
        display_data(df3)

    elif pilihan == '4':
        print("Terima kasih. Program selesai.")
        break

    else:
        print("Pilihan tidak valid. Silakan pilih menu yang tersedia.")
```
