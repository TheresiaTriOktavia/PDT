```python
# Menghapus kolom non numerik
df_clean2 = df2.drop(["Country", "Status"], axis=1)
df_clean2
```

```python
# Menghitung jumlah (sum) pada tiap kolom
# Mengatur sehingga hanya ada 2 angka di belakang koma
DataSum2 = df_clean2.sum().apply(lambda x: "{:.2f}".format(x))
DataSum2
```

```python
# Menghitung rata-rata (mean) pada tiap kolom
# Mengatur sehingga hanya ada 2 angka di belakang koma
DataMean2 = df_clean2.mean().apply(lambda x: "{:.2f}".format(x))
DataMean2
```

```python
# Menghitung nilai tengah (median) pada tiap kolom
# Mengatur sehingga hanya ada 2 angka di belakang koma
DataMedian2 = df_clean2.median().apply(lambda x: "{:.2f}".format(x))
DataMedian2
```
